```python
import pandas as pd

# Load CSV file
df = pd.read_csv("regulations.csv")

print(df)

```

```python
df = df.iloc[0:517, :3]
```

```python

# Shift column 2 down by 1 row
df['col2_shifted'] = df['1'].shift(1)

# Concatenate col1 and shifted col2 (handling NaN values)
df['0'] = df['0'] + '-' + df['col2_shifted'].fillna('')

# Drop the temporary column if not needed
df.drop(columns=['col2_shifted'], inplace=True)
```

```python
df = df.drop(columns=['1'])
```

```python
df = df.dropna(how='all')
```

```python
df = df.rename(columns={'0': 'Standard', '2': 'Guidance'})
```

```python
df.to_csv('RegulationsFiltered.csv', index=False)
```
